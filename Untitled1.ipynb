```python
# ===============================
# 1. Import Libraries
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# ===============================
# 2. Load Dataset
# ===============================
# Upload your CSV file in Colab and use its name here
data = pd.read_excel('Concrete_Data.xls')

print(data.head())

# Rename columns (optional for easy use)
data.columns = ['cement', 'slag', 'flyash', 'water', 'superplasticizer',
                'coarseagg', 'fineagg', 'age', 'strength']

# ===============================
# 3. EDA (Exploratory Data Analysis)
# ===============================

# Strength distribution
plt.hist(data['strength'], bins=20) # Added bins argument
plt.title("Strength Distribution")
plt.xlabel("Strength (MPa)")
plt.ylabel("Frequency")
plt.show()

# Cement vs Strength
plt.scatter(data['cement'], data['strength'])
plt.xlabel("Cement")
plt.ylabel("Strength")
plt.title("Cement vs Strength")
plt.show()

# Age vs Strength
plt.scatter(data['age'], data['strength'])
plt.xlabel("Age")
plt.ylabel("Strength")
plt.title("Age vs Strength")
plt.show()

# ===============================
# 4. Data Preprocessing
# ===============================

X = data.drop('strength', axis=1)
y = data['strength']

# Min-Max Scaling
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# ===============================
# 5. Linear Regression Model
# ===============================
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
r2_lr = r2_score(y_test, y_pred_lr)

print("Linear Regression RMSE:", rmse_lr)
print("Linear Regression R2:", r2_lr)

# ===============================
# 6. Polynomial Regression (degree=2)
# ===============================
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_scaled)

X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(X_poly, y, test_size=0.2, random_state=42)

lr_poly = LinearRegression()
lr_poly.fit(X_train_p, y_train_p)

y_pred_poly = lr_poly.predict(X_test_p)

rmse_poly = np.sqrt(mean_squared_error(y_test_p, y_pred_poly))
r2_poly = r2_score(y_test_p, y_pred_poly)

print("Polynomial RMSE:", rmse_poly)
print("Polynomial R2:", r2_poly)

# ===============================
# 7. Plot: Strength vs Age
# ===============================
plt.scatter(data['age'], data['strength'])
plt.xlabel("Age")
plt.ylabel("Strength")
plt.title("Strength vs Age")
plt.show()

# ===============================
# 8. Residual Plot
# ===============================
residuals = y_test - y_pred_lr

plt.scatter(y_pred_lr, residuals)
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.axhline(y=0)
plt.show()

# ===============================
# 9. Feature Importance (Coefficients)
# ===============================
coeff_df = pd.DataFrame(lr.coef_, X.columns, columns=['Coefficient'])
print(coeff_df)

print("\nMost Important Feature:")
print(coeff_df.sort_values(by='Coefficient', ascending=False).head(1))
```
